```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread("/content/image1.jpg",0)
img2=cv2.imread("/content/image3.jpg",0)

clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
clahe_img=clahe.apply(img1)


clahe1=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
clahe1_img=clahe1.apply(img2)



```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img1,cmap='bone_r')
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Clahe image")
plt.imshow(clahe_img,cmap='gist_heat')
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img2,cmap='Purples')
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Clahe image")
plt.imshow(clahe1_img,cmap='gray')
plt.axis("off")
plt.show()
```
