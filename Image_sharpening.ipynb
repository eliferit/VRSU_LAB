```python
import cv2
import matplotlib as pl
import numpy as np
```

```python
img=cv2.imread("/content/image4.jpg",0)
```

```python
kernel=np.array([[0,-1,0],
                 [-1,5,-1],
                 [0,-1,0]])
```

```python
sharpned=cv2.filter2D(img,-1,kernel)
```

```python
pl.figure(figsize=(10,5))
pl.subplot(1,2,1)
pl.imshow(img,cmap='gray')
pl.title("Original")
pl.axis('off')
pl.subplot(1,2,2)
pl.imshow(sharpned,cmap='gray')
pl.axis('off')
```
