```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("/content/assortment-leaves-flowers-dark-background.jpg",0)
```

```python
he_img=cv2.equalizeHist(img)
```

```python
plt.figure(figsize=(12,5))
```

```python
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img,cmap='gnuplot_r')
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Histogram")
plt.imshow(he_img,cmap='gnuplot_r')
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Histogram")
plt.hist(he_img.ravel(),bins=256)
plt.show()

```
